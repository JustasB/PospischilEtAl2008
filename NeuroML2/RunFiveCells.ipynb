## Import pyNeuroML & run cell in jNeuroML_NEURON

```python
from pyneuroml import pynml



lems_file = 'LEMS_FiveCells.xml'

results = pynml.run_lems_with_jneuroml_neuron(lems_file, nogui=True, load_saved_data=True)

```

## Reload data and plot

```python
from matplotlib import pyplot as plt

for key in results.keys():
    plt.xlabel('Time (ms)')
    plt.ylabel('Membrane potential (mV)')

    if key != 't':
        plt.plot([t*1000. for t in results['t']],[v*1000. for v in results[key]], label="Cell %s"%key.split('/')[2])
        
        plt.legend(loc=5)
        plt.figure()
    
plt.show()
    
```
